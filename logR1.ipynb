```python
edu_mean = heartDis.education.mean()
glu_mean = heartDis.glucose.mean()
heartDis.education.fillna(edu_mean, axis =0, inplace = True)
heartDis.glucose.fillna(glu_mean, axis = 0, inplace = True)

#fillna fills the nan values with the mean of the columns
```

```python
heartDis.dropna(inplace = True)
```

```python
from sklearn.model_selection import train_test_split
heartDis.columns
#getting the columns to copy paste
```

```python
X = heartDis[['male', 'age', 'education', 'currentSmoker', 'cigsPerDay', 'BPMeds',
       'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP',
       'diaBP', 'BMI', 'heartRate', 'glucose']]
#all of the variables used to predict CHD

y = heartDis["TenYearCHD"]
#what you're predicting

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#splitting the data into training and testing
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression()
logModel.fit(X_train, y_train)
```

```python
predictions = logModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
logModel.score(X_test, y_test)
#accuracy of the model
```
